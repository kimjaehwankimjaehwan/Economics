###연방기금 유효이율이란?
연방기금 유효이율은 예금 기관(은행이나 신용 조합 등)이 다른 예금 기관에게 하룻밤 동안 초과 준비금을 대출해 줄 때 적용되는 금리입니다. "유효"라는 단어는 실제로 거래된 이자율을 나타내며, 연방준비제도(Federal Reserve, Fed)에서 설정한 목표 금리와는 다소 차이가 있을 수 있습니다.

###작동 원리:
1. 연방기금 시장:

미국의 은행들은 고객의 예금 인출 요청이나 기타 의무를 이행할 수 있도록 연방준비제도에 일정량의 준비금을 보유해야 합니다.
어떤 은행들은 초과 준비금을 보유하고 있는 반면, 다른 은행들은 준비금을 충족하기 위해 대출을 필요로 할 수 있습니다. 이때 은행들이 하룻밤 동안 준비금을 빌려주고 빌리는 시장이 연방기금 시장입니다.

2. 금리 결정:

연방기금 유효이율은 연방기금 시장에서 준비금에 대한 수요와 공급에 따라 결정됩니다. 이율은 이러한 시장 조건에 따라 매일 변동됩니다.
연방준비제도는 연방기금 목표 금리 범위를 설정하지만, 실제 유효이율은 시장 역학에 따라 이 범위 내에서 변동할 수 있습니다.

3. 연방준비제도의 역할:

연방준비제도는 공개 시장 조작을 통해 연방기금 유효이율에 영향을 미칩니다. 예를 들어, 정부 증권을 매수하거나 매도하여 은행 시스템 내 준비금의 양을 증가시키거나 감소시킵니다.
이를 통해 연방준비제도는 유효이율을 목표 범위로 유도할 수 있습니다.

##Federal Funds Rate 변화의 경제적 영향

###금융 시장에 미치는 영향:

  - 채권 시장: Federal Funds Rate가 상승하면 채권 수익률이 상승하고, 채권 가격이 하락하는 경향이 있습니다. 반대로 금리가 하락하면 채권 수익률이 하락하고 채권 가격은 상승합니다.

  - 주식 시장: 금리 상승은 기업의 자본 비용을 증가시켜 주식 시장에 부정적인 영향을 줄 수 있습니다. 반면 금리 인하는 기업의 대출 비용을 낮추고, 주식 시장에 긍정적인 영향을 미칠 수 있습니다.
  
  - 외환 시장: 금리 상승은 달러화를 강세로 만들 수 있으며, 금리 하락은 달러화를 약세로 만들 수 있습니다. 이는 다른 국가와의 금리 차이로 인해 자본 흐름에 영향을 미치기 때문입니다.

###소비자와 기업에 미치는 영향:

  - 소비자 대출: 금리가 상승하면 주택담보대출, 자동차 대출, 신용카드 이자율이 상승하여 소비자 지출이 감소할 수 있습니다.
  - 기업 대출: 금리 인상은 기업의 차입 비용을 높여 투자를 억제할 수 있으며, 이는 경제 성장 둔화로 이어질 수 있습니다.

###인플레이션과 실업률:

  - 인플레이션 통제: Federal Funds Rate가 상승하면 차입 비용이 증가하고 소비가 감소하여 인플레이션을 억제할 수 있습니다.
  - 실업률: 금리 상승은 경제 활동을 둔화시켜 실업률을 상승시킬 수 있으며, 반대로 금리 인하는 경제 활동을 촉진하여 실업률을 낮출 수 있습니다.

###경제 성장:

  - GDP 성장: 금리 하락은 소비와 투자를 증가시켜 GDP 성장을 촉진할 수 있습니다. 반대로 금리 인상은 경제 성장을 둔화시킬 수 있습니다.

```python
inflation_file_path = '/content/Inflation, consumer prices.xlsx'
inflation_df = pd.read_excel(inflation_file_path)

# Extract US data
us_inflation_df = inflation_df[inflation_df['Country Name'] == 'United States']

# Clean and reshape the inflation data
us_inflation_df = us_inflation_df.melt(id_vars=['Country Name', 'Series Name', 'Series Code', 'Country Code'],
                                       var_name='Year', value_name='Inflation Rate')

# Clean up the 'Year' column
us_inflation_df['Year'] = us_inflation_df['Year'].str.extract('(\d{4})').astype(int)

# Load the Federal Funds Effective Rate data
federal_funds_rate_file_path = '/content/Federal Funds Effective Rate_yearly.xlsx'
ffr_df = pd.read_excel(federal_funds_rate_file_path)

# Convert 'observation_date' to datetime and extract the year
ffr_df['observation_date'] = pd.to_datetime(ffr_df['observation_date'])
ffr_df['Year'] = ffr_df['observation_date'].dt.year # Extract year as integer to match 'Year' in us_inflation_df

# Merge the datasets on 'Year'
merged_df = pd.merge(ffr_df, us_inflation_df[['Year', 'Inflation Rate']], on='Year') # Merge on the now common 'Year' column

# Drop redundant columns
merged_df = merged_df.drop(columns=['observation_date']) # Drop the original datetime column

# Rename columns for clarity
merged_df.rename(columns={'Federal Funds Effective Rate': 'Federal Funds Rate'}, inplace=True)

# Explicitly convert 'Inflation Rate' to numeric, coercing errors
merged_df['Inflation Rate'] = pd.to_numeric(merged_df['Inflation Rate'], errors='coerce')

# Display the merged dataset
print(merged_df)

```

### VAR 분석(inflation, federal rate)

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from sklearn.preprocessing import StandardScaler

# 데이터 로드 (앞에서 사용한 Federal Funds Rate와 인플레이션 데이터)
# 여기에 Merge된 데이터프레임 merged_df를 사용

# VAR 모델에 적합한 형태로 데이터 준비
# 필요하다면 데이터가 안정적(stationary)인지 확인하고 차분(differencing)을 수행합니다.
# 만약 차분이 필요하면 다음과 같은 코드를 사용할 수 있습니다:
# merged_df['Federal Funds Rate Diff'] = merged_df['Federal Funds Rate'].diff().dropna()
# merged_df['Inflation Rate Diff'] = merged_df['Inflation Rate'].diff().dropna()

# 안정성 테스트 (Augmented Dickey-Fuller Test)
def adf_test(series, title=''):
    """
    Augmented Dickey-Fuller Test
    Null Hypothesis: Data is non-stationary
    """
    result = adfuller(series.dropna(), autolag='AIC')
    labels = ['ADF Test Statistic', 'p-value', '# Lags Used', '# Observations Used']
    out = pd.Series(result[0:4], index=labels)
    for key, value in result[4].items():
        out[f'Critical Value ({key})'] = value
    print(f'Augmented Dickey-Fuller Test on "{title}"\n')
    print(out.to_string())
    print('\n')

# ADF Test on the original data
adf_test(merged_df['Federal Funds Rate'], 'Federal Funds Rate')
adf_test(merged_df['Inflation Rate'], 'Inflation Rate')

# 데이터가 안정적이지 않다면, 차분을 통해 안정화합니다.
merged_df_diff = merged_df.diff().dropna()

# ADF Test on the differenced data
adf_test(merged_df_diff['Federal Funds Rate'], 'Differenced Federal Funds Rate')
adf_test(merged_df_diff['Inflation Rate'], 'Differenced Inflation Rate')
```

  - Critical Values:
    - 1%: -4.012034
    - 5%: -3.104184
    - 10%: -2.690987
  
  - 해석:
    - ADF Test Statistic이 1%, 5%, 10%의 임계값보다 큽니다(즉, 덜 음수). p-value가 0.096으로 0.05보다 크므로 귀무가설을 기각할 수 없습니다.
    - 이는 데이터가 안정적이지 않다는 것을 의미합니다. 따라서 차분이 필요합니다.

    - 이 결과는 데이터가 여전히 약간 불안정할 수 있음을 시사하지만, 일반적으로 이 정도면 모델링에 사용할 수 있는 수준으로 간주될 수 있습니다.

    - 이는 데이터가 차분 후 안정적(stationary)임을 강하게 시사합니다.

L1.Inflation Rate의 계수는 -0.661324이며, 이는 인플레이션율이 과거에 있었던 변화를 통해 현재 인플레이션율에 부정적인 영향을 미친다는 것을 나타냅니다(p-값 < 0.05).

```python
# 충격 반응 함수 (Impulse Response Function, IRF)
irf = var_model.irf(10)
irf.plot(orth=False)
plt.show()

# 예측 오차 분해 (Forecast Error Variance Decomposition, FEVD)
fevd = var_model.fevd(10)
fevd.plot(figsize=(10, 6))
plt.show()

# 모델 안정성 테스트
stability = var_model.is_stable()
print(f"Model Stability: {'Stable' if stability else 'Unstable'}")
```

### Impulse Response Function, IRF
  - 각 그래프의 의미:

    - 그래프 제목은 충격을 가한 변수와 반응을 측정한 변수를 나타냅니다. 예를 들어, "Federal Funds Rate → Inflation Rate"은 Federal Funds Rate에 충격을 가했을 때 Inflation Rate의 반응을 보여줍니다.

    - 파란색 실선은 충격에 대한 반응을 나타내며, 점선은 신뢰 구간을 나타냅니다. 신뢰 구간이 0을 포함하면 반응이 통계적으로 유의하지 않을 수 있습니다.

  - 중요한 관찰점:

    - Federal Funds Rate → Federal Funds Rate: Federal Funds Rate에 충격을 가했을 때, Federal Funds Rate 자체가 빠르게 하락한 후 점진적으로 원래 상태로 회복됩니다.
    - Federal Funds Rate → Inflation Rate: Federal Funds Rate에 충격을 가했을 때 Inflation Rate도 반응을 보이지만, 그 반응은 상대적으로 작고 신뢰 구간이 넓습니다.
    - Inflation Rate → Federal Funds Rate Change: Inflation Rate에 충격을 가했을 때 Federal Funds Rate Change가 즉각적으로 반응하지만, 그 영향은 점차 감소합니다.
    - Federal Funds Rate Change → Inflation Rate: Federal Funds Rate의 변화가 Inflation Rate에 즉각적으로 영향을 미치며, 그 영향은 시간이 지남에 따라 점차 약해집니다.

  - 정책적 해석:

    - 이러한 충격 반응 함수는 경제 정책의 효과를 분석하는 데 유용합니다. 예를 들어, 중앙은행이 금리를 인상했을 때 인플레이션에 미치는 영향의 크기와 지속 시간을 분석할 수 있습니다.
    - 금융 시장이나 다른 경제 변수에서 정책 충격이 어떻게 반응하는지를 파악하여, 정책의 기대 효과와 실제 효과를 비교할 수 있습니다.


### 예측 오차 분해(Forecast Error Variance Decomposition, FEVD)

  - 각 변수에 대한 예측 오차 분해:

    - Federal Funds Rate:

      - Federal Funds Rate의 예측 오차는 주로 자신에 의해 설명됩니다. 이는 Federal Funds Rate가 강한 자기상관성을 가지고 있음을 시사합니다.
      - 시간이 지남에 따라 Inflation Rate와 다른 변수들이 Federal Funds Rate 예측 오차의 일부를 설명하기 시작하지만, 그 영향은 미미합니다.

    - Inflation Rate:

      - 초기에는 Inflation Rate의 예측 오차가 주로 자신에 의해 설명됩니다.
      - 시간이 지남에 따라 Federal Funds Rate와 그 변화가 Inflation Rate 예측 오차의 일부를 설명하게 됩니다. 이는 금리가 인플레이션에 미치는 영향을 나타낼 수 있습니다.

    - Federal Funds Rate Change:

      - Federal Funds Rate Change의 예측 오차는 시간이 지남에 따라 주로 Federal Funds Rate와 그 자체의 변화에 의해 설명됩니다.
      - Inflation Rate의 영향은 점차 증가하지만, 여전히 주요한 설명 변수는 Federal Funds Rate입니다.

    - Inflation Rate Change:

      - Inflation Rate Change의 예측 오차는 초기에 주로 자신에 의해 설명되며, 시간이 지남에 따라 Federal Funds Rate의 영향이 커집니다.
  
  - 정책적 해석:

    - Federal Funds Rate와 그 변화가 인플레이션과 그 변화에 중요한 영향을 미치고 있음을 시사합니다.
    - 시간이 지남에 따라 인플레이션과 그 변화가 Federal Funds Rate 및 그 변화에 미치는 영향은 상대적으로 작습니다. 이는 금리가 인플레이션 조정에서 더 중요한 역할을 할 수 있음을 암시합니다.
