##Inflation: Progress and the Path Ahead

작년 잭슨홀 심포지엄에서 저는 간결하고 직접적인 메시지를 전달했습니다.

올해의 발언은 조금 더 길지만, 메시지는 동일합니다:

**연준의 임무는 인플레이션을 2% 목표로 낮추는 것이며, 우리는 그렇게 할 것입니다.**
지난해 동안 우리는 정책을 상당히 긴축했습니다.

인플레이션이 최고점에서 하락한 것은 환영할 만한 발전이지만, 여전히 너무 높습니다.

우리는 필요하다면 추가로 금리를 인상할 준비가 되어 있으며, 인플레이션이 목표를 향해 지속적으로 하락하고 있다고 확신할 때까지 정책을 제한적인 수준으로 유지할 계획입니다.

오늘 저는 지금까지의 진전을 검토하고, 우리의 이중 목표를 추구하는 과정에서 우리가 직면한 전망과 불확실성에 대해 논의할 것입니다.

마지막으로, 이것이 정책에 의미하는 바를 요약하겠습니다.

우리가 지금까지 이루어온 만큼, 다가오는 회의에서 우리는 들어오는 데이터와 변화하는 전망과 위험을 평가하면서 신중히 진행할 위치에 있습니다.

**지금까지의 인플레이션 감소**

현재의 높은 인플레이션 현상은 매우 강한 수요와 팬데믹으로 인한 공급 제약 간의 충돌에서 처음 발생했습니다.

연방공개시장위원회(FOMC)가 2022년 3월에 정책 금리를 인상했을 때, 인플레이션을 낮추는 것은 팬데믹 관련 수요와 공급 왜곡의 전례 없는 해소와 우리의 통화정책 긴축이 함께 작용하여 총수요 증가를 둔화시키고 공급이 따라잡을 시간을 제공하는 데 달려있다는 것이 명확했습니다.

이제 이 두 가지 힘이 인플레이션을 낮추기 위해 함께 작용하고 있지만, 더 최근의 긍정적인 수치에도 불구하고 이 과정은 아직 갈 길이 멉니다.

12개월 기준으로 미국의 전체 또는 '헤드라인' PCE(개인소비지출) 인플레이션은 2022년 6월에 7%로 정점을 찍었고, 이후 7월에는 3.3%로 하락했습니다(그림 1, 패널 A).

러시아의 우크라이나 전쟁은 2022년 초 이후 전 세계 헤드라인 인플레이션 변화의 주요 원인이었습니다.

헤드라인 인플레이션은 가계와 기업이 가장 직접적으로 경험하는 것이기 때문에, 이 하락은 매우 좋은 소식입니다.

하지만 식품과 에너지 가격은 여전히 전 세계적인 요인에 의해 영향을 받아 변동성이 크며, 인플레이션이 어디로 향하고 있는지에 대한 잘못된 신호를 줄 수 있습니다.

남은 발언에서 저는 식품과 에너지 요소를 제외한 근원 PCE 인플레이션에 초점을 맞출 것입니다.

12개월 기준으로 근원 PCE 인플레이션은 2022년 2월에 5.4%로 정점을 찍었고, 7월에는 4.3%로 점진적으로 하락했습니다(그림 1, 패널 B).

6월과 7월의 근원 인플레이션에 대한 낮은 월별 수치는 환영할 만하지만, 좋은 데이터가 두 달 동안 나온 것만으로는 인플레이션이 목표를 향해 지속적으로 하락하고 있다는 자신감을 구축하기에는 부족합니다.

이 낮은 수치들이 계속될지, 근본적인 인플레이션이 앞으로 몇 분기에 걸쳐 어디에 자리 잡을지는 아직 알 수 없습니다.

12개월 근원 인플레이션은 여전히 높은 수준을 유지하고 있으며, 가격 안정성으로 돌아가기 위해서는 상당한 추가 진전이 필요합니다.

향후 진전을 주도할 가능성이 있는 요인을 이해하기 위해서는 근원 PCE 인플레이션의 세 가지 광범위한 구성 요소인 상품, 주거 서비스, 기타 모든 서비스(때로는 비주거 서비스라고도 함)를 개별적으로 검토하는 것이 유용합니다(그림 2).

근원 상품 인플레이션은 특히 내구재의 경우 급격히 하락했습니다.
이는 긴축된 통화 정책과 공급 및 수요 불균형의 느린 해소가 인플레이션을 낮추고 있기 때문입니다.

자동차 부문이 좋은 예입니다.
팬데믹 초기에는 저금리, 재정 지원, 대면 서비스 지출 감소, 대중교통 사용 및 도시 생활 선호도의 변화로 인해 차량 수요가 급격히 증가했습니다.
하지만 반도체 부족으로 인해 차량 공급이 실제로 감소했습니다.
차량 가격이 급등했고, 큰 규모의 억눌린 수요가 생겨났습니다.
팬데믹과 그 영향이 줄어들면서 생산과 재고가 증가하고 공급이 개선되었습니다.
동시에, 높은 금리가 수요에 부담을 주었습니다.
자동차 대출 금리는 지난해 초보다 거의 두 배로 증가했으며, 고객들은 높은 금리가 차량 구매력에 미치는 영향을 느끼고 있다고 보고합니다.
결과적으로, 공급과 수요 요인의 결합 효과로 인해 자동차 인플레이션이 급격히 감소했습니다.

이와 유사한 역학이 근원 상품 인플레이션 전반에서 일어나고 있습니다.
이러한 현상이 진행됨에 따라 통화 제약의 효과가 시간이 지남에 따라 더 명확히 나타날 것입니다.
근원 상품 가격은 지난 두 달 동안 하락했지만, 12개월 기준으로는 여전히 팬데믹 이전 수준을 훨씬 상회하고 있습니다.
지속적인 진전이 필요하며, 이를 달성하기 위해서는 제한적인 통화 정책이 요구됩니다.

금리에 민감한 주택 부문에서는 정책 효과가 조기부터 나타났습니다.
2022년 동안 주택담보대출 금리가 두 배로 증가하여 주택 착공과 매매가 감소하고 주택 가격 상승이 급락했습니다.
시장 임대료의 상승세는 곧 정점을 찍고 점진적으로 하락하기 시작했습니다(그림 3).

측정된 주거 서비스 인플레이션은 이러한 변화에 뒤늦게 반영되었지만, 최근에는 하락하기 시작했습니다.
이 인플레이션 지표는 모든 세입자가 지불한 임대료뿐만 아니라, 자가 거주 주택에서 얻을 수 있는 임대료에 대한 추정치도 반영합니다.
임대 계약이 천천히 갱신되기 때문에 시장 임대료 상승률의 감소가 전체 인플레이션 지표에 반영되기까지 시간이 걸립니다.
지난 1년 동안 새로운 임대 계약의 임대료 상승세 둔화는 '파이프라인'에 있는 것으로 생각할 수 있으며, 앞으로 1년 동안 측정된 주거 서비스 인플레이션에 영향을 미칠 것입니다.
앞으로 시장 임대료 상승이 팬데믹 이전 수준에 가까워지면, 주거 서비스 인플레이션도 팬데믹 이전 수준으로 하락할 것입니다.
우리는 주거 서비스 인플레이션에 대한 상하 위험의 신호를 확인하기 위해 시장 임대료 데이터를 계속 주시할 것입니다.

마지막으로, 비주거 서비스는 근원 PCE 지수의 절반 이상을 차지하며, 의료 서비스, 식품 서비스, 교통 및 숙박 등을 포함하는 광범위한 서비스를 포함합니다.
이 부문에서 12개월 인플레이션은 금리 인상 이후 거의 변하지 않았습니다.
하지만 최근 3개월과 6개월 동안 측정된 인플레이션은 하락하고 있어 고무적입니다.
비주거 서비스 인플레이션의 소폭 하락의 부분적인 이유는 많은 이러한 서비스가 글로벌 공급망 병목 현상에 덜 영향을 받았으며, 주택이나 내구재와 같은 다른 부문보다 금리에 덜 민감한 것으로 여겨지기 때문입니다.
이러한 서비스의 생산은 또한 상대적으로 노동 집약적이며, 노동 시장은 여전히 타이트합니다.
이 부문의 규모를 고려할 때, 가격 안정성을 회복하기 위해서는 이 부문에서도 추가적인 진전이 필수적입니다.
시간이 지남에 따라, 제한적인 통화 정책은 총 공급과 수요를 더 잘 균형 잡아 인플레이션 압력을 감소시킬 것입니다.


**전망**

전망으로 전환하면, 팬데믹 관련 왜곡의 추가적인 해소는 계속해서 인플레이션에 하방 압력을 가할 가능성이 있지만, 제한적인 통화 정책이 점점 더 중요한 역할을 할 가능성이 큽니다.
인플레이션을 지속적으로 2%로 낮추기 위해서는 경제 성장이 잠재성장률 이하로 둔화되고, 노동 시장 여건이 다소 완화되는 기간이 필요할 것으로 예상됩니다.


**경제 성장**

제한적인 통화 정책은 금융 조건을 긴축시켰으며, 이는 잠재성장률 이하의 성장을 기대하는 데 기여했습니다.
작년 심포지엄 이후로 2년물 실질 금리는 약 250bp 상승했으며, 장기 실질 금리도 약 150bp 상승했습니다.
금리 변화 외에도 은행 대출 기준이 강화되었고, 대출 성장률이 급격히 둔화되었습니다.
이러한 광범위한 금융 조건의 긴축은 일반적으로 경제 활동 성장의 둔화에 기여하며, 이번 사이클에서도 그 증거를 볼 수 있습니다.
예를 들어, 산업 생산의 성장은 둔화되었으며, 주거 투자에 지출된 금액은 지난 5분기 동안 각각 감소했습니다(그림 4).

그러나 우리는 경제가 예상대로 냉각되지 않을 수 있다는 신호에 주목하고 있습니다.
올해 들어 GDP(국내총생산) 성장은 예상보다 높았으며, 최근의 소비자 지출 수치도 특히 강세를 보였습니다.
또한 지난 18개월 동안 급격히 둔화된 후, 주택 부문이 회복 조짐을 보이고 있습니다.
지속적인 잠재성장률 이상의 성장이 지속되면 인플레이션에 대한 진전을 위험에 빠뜨릴 수 있으며, 통화 정책의 추가적인 긴축이 필요할 수 있습니다.

**노동 시장**

노동 시장의 재균형이 지난 1년 동안 계속되었지만, 여전히 완전하지 않습니다.
노동 공급은 25~54세 근로자의 참여가 강화되고 팬데믹 이전 수준으로의 이민 증가에 의해 개선되었습니다.
실제로, 주요 근로 연령의 여성의 노동 참여율은 6월에 사상 최고치를 기록했습니다.
노동에 대한 수요도 완화되었습니다.
구인 공고는 여전히 많지만, 감소 추세에 있습니다.
고용 증가율은 크게 둔화되었습니다.
총 노동 시간은 지난 6개월 동안 변하지 않았으며, 주당 평균 근로 시간은 팬데믹 이전 범위의 하단으로 감소하여 노동 시장 조건의 점진적인 정상화를 반영하고 있습니다(그림 5).

이 재균형은 임금 압력을 완화시켰습니다.
여러 지표에 걸쳐 임금 성장은 점진적으로 둔화되고 있습니다(그림 6).
명목 임금 성장이 궁극적으로 2% 인플레이션과 일치하는 수준으로 둔화되어야 하지만, 가계에 중요한 것은 실질 임금 성장입니다.
명목 임금 성장이 둔화되었음에도 불구하고, 인플레이션이 하락함에 따라 실질 임금 성장은 증가하고 있습니다.

우리는 이 노동 시장 재균형이 계속될 것으로 예상합니다.
노동 시장의 타이트함이 더 이상 완화되지 않는다는 증거는 또한 통화 정책 대응이 필요할 수 있습니다.

**앞으로의 경로에서의 불확실성 및 리스크 관리**

2%는 우리 인플레이션 목표이며, 그 목표를 달성하고 유지하는 것이 우리의 임무입니다.
우리는 시간이 지남에 따라 인플레이션을 그 수준으로 낮추기 위해 충분히 제한적인 통화 정책을 달성하고 유지하기 위해 노력하고 있습니다.
물론, 실시간으로 그런 입장을 달성했는지 아는 것은 어렵습니다.
모든 긴축 사이클에서 공통적인 몇 가지 도전 과제가 있습니다.
예를 들어, 실질 금리는 이제 양성이며, 중립 정책 금리에 대한 주류 추정치를 훨씬 상회하고 있습니다.
우리는 현재의 정책 입장을 제한적으로 보고 있으며, 경제 활동, 고용, 인플레이션에 하방 압력을 가하고 있습니다.
그러나 우리는 중립 금리를 정확히 식별할 수 없으므로, 통화 정책의 제한 수준에 대한 불확실성이 항상 존재합니다.

이러한 평가는 통화 긴축이 경제 활동과 특히 인플레이션에 영향을 미치는 지연 기간에 대한 불확실성으로 인해 더욱 복잡해집니다.
작년 심포지엄 이후, 위원회는 정책 금리를 300bp 인상했으며, 그 중 지난 7개월 동안 100bp를 인상했습니다.
그리고 우리는 증권 보유량을 상당히 줄였습니다.

이러한 전통적인 정책 불확실성 외에도, 이번 사이클에서의 공급 및 수요 왜곡은 인플레이션과 노동 시장 역학에 영향을 미치면서 추가적인 복잡성을 초래합니다.
예를 들어, 지금까지 구인 공고는 실업률 증가 없이 크게 감소했습니다.
이는 매우 환영할 만한 일이지만, 역사적으로는 이례적인 결과이며, 노동 수요의 과잉을 반영하는 것처럼 보입니다.
또한, 인플레이션이 최근 몇 십년보다 노동 시장 타이트함에 더 민감하게 반응했다는 증거가 있습니다.
이러한 변화하는 역학이 지속될지 여부는 불확실하며, 이 불확실성은 민첩한 정책 결정의 필요성을 강조합니다.

이러한 오래된 불확실성과 새로운 불확실성은 통화 정책을 너무 많이 긴축하는 것과 너무 적게 긴축하는 것의 위험을 균형 맞추는 우리의 과제를 복잡하게 만듭니다.
