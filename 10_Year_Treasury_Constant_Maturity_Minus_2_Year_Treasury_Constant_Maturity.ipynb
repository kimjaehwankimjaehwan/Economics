##수익률 곡선 스프레드

  - 10년 만기 미국 국채와 2년 만기 미국 국채의 수익률 차이 : 경제 상황을 평가하는 중요한 지표로, 특히 경기 침체의 신호로 많이 사용

  - 경제적 해석
    - 양의 스프레드: 10년 국채 수익률이 2년 국채 수익률보다 높은 경우, 장기 경제 전망이 낙관적이라는 신호로 해석될 수 있습니다.

    - 음의 스프레드 (역전): 2년 국채 수익률이 10년 국채 수익률보다 높은 경우, 경제 침체의 가능성을 나타내는 중요한 신호일 수 있습니다.

```python
import pandas as pd

data = pd.read_excel('/content/T10Y2Y.xls', sheet_name='FRED Graph')

print(data.head())
```

```python
data.dropna(inplace=True)

data['observation_date'] = pd.to_datetime(data['observation_date'])

print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data['observation_date'], data['T10Y2Y'], label='10-Year Minus 2-Year Treasury Yield Spread')
plt.axhline(0, color='red', linestyle='--', label='Zero Spread')
plt.title('10-Year Minus 2-Year Treasury Yield Spread Over Time')
plt.xlabel('Date')
plt.ylabel('Yield Spread (%)')
plt.legend()
plt.grid(True)
plt.show()
```

###그래프 해석:
  - 양의 스프레드:

    - 2020년 초부터 2021년 중반까지 스프레드가 긍정적이었으며, 이는 10년 국채의 수익률이 2년 국채의 수익률보다 높았다는 것을 의미합니다. 이는 장기적으로 경제 성장에 대한 낙관적인 기대를 반영할 수 있습니다.

  - 음의 스프레드 (수익률 곡선 역전):

    - 2022년 중반 이후, 스프레드가 음수로 전환되는 시점이 보입니다. 이는 역사적으로 경기 침체의 신호로 해석될 수 있습니다. 예를 들어, 수익률 곡선의 역전은 과거 경기 침체를 예측하는 데 중요한 지표로 사용되었습니다.

  - 최근의 회복:

    - 2024년 초부터 스프레드가 다시 상승하는 경향이 보입니다. 이는 경제가 회복될 가능성을 시사할 수 있습니다. 하지만 스프레드가 아직 낮은 수준에 머물러 있는지, 아니면 더 높은 수준으로 회복될지 지켜볼 필요가 있습니다.

```python
print("Basic Statistics of the Yield Spread:")
print(data['T10Y2Y'].describe())

min_spread = data['T10Y2Y'].min()
max_spread = data['T10Y2Y'].max()
mean_spread = data['T10Y2Y'].mean()

print(f"\nMinimum Spread: {min_spread}%")
print(f"Maximum Spread: {max_spread}%")
print(f"Average Spread: {mean_spread}%")
```

###통계 해석:
  - 평균 스프레드 (Mean Spread):

    - 평균 스프레드는 약 **0.1756%**입니다. 이는 일반적으로 10년 만기 국채 수익률이 2년 만기 국채 수익률보다 약간 더 높았음을 의미합니다. 하지만 평균 값이 매우 낮기 때문에, 장단기 금리 차이가 전반적으로 크지 않았음을 알 수 있습니다.

  - 표준편차 (Standard Deviation):

    - 표준편차는 약 **0.6765%**입니다. 이는 스프레드의 변동성이 비교적 크다는 것을 나타냅니다. 즉, 경제 상황에 따라 10년과 2년 만기 국채 간의 수익률 차이가 크게 변동했음을 시사합니다.

  - 최소값 (Minimum Spread):

    - 최소 스프레드는 **-1.08%**입니다. 이 값은 2년 만기 국채 수익률이 10년 만기 국채 수익률보다 1.08% 높았던 시점을 의미하며, 이는 상당히 큰 수익률 곡선의 역전을 나타냅니다. 이런 상황은 보통 경기 침체의 신호로 해석됩니다.

  - 최대값 (Maximum Spread):

    - 최대 스프레드는 **1.59%**입니다. 이는 10년 만기 국채 수익률이 2년 만기 국채 수익률보다 1.59% 높았던 시점을 의미합니다. 이는 경제에 대한 장기적 낙관주의를 나타낼 수 있는 시점입니다.

  - 중앙값 (Median):

    - 중앙값은 **0.18%**로, 데이터의 중간 지점을 나타냅니다. 중앙값이 평균과 비슷한 수준에 있다는 것은 데이터가 상대적으로 균형 있게 분포되어 있음을 시사합니다.

  - 사분위수 (Quartiles):

    - 25% 사분위수는 -0.39%, 75% 사분위수는 **0.67%**입니다. 이는 데이터가 아래쪽으로 약간 치우쳐 있음을 의미하며, 스프레드가 음수였던 기간이 꽤 있었음을 나타냅니다.
