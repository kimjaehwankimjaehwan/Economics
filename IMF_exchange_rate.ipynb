```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/IMF_exchange_rate.xlsx'
df = pd.read_excel(file_path)

df = pd.DataFrame(df)
df.head()
```

```python
df.columns
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df.columns = ['Chinese Yuan', 'EURO', 'Korean Won', 'UK Pound', 'US Dollar']
```

```python
df
```

```python
df_filled = df.interpolate(method='linear', axis=0)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

currencies = ['Chinese Yuan', 'EURO', 'Korean Won', 'UK Pound', 'US Dollar']
colors = ['blue', 'orange', 'green', 'red', 'purple']

fig, axes = plt.subplots(nrows=len(currencies), ncols=1, figsize=(12, 8 * len(currencies)))

for ax, currency, color in zip(axes, currencies, colors):
    ax.plot(df_filled.index, df_filled[currency], color=color, label=currency)
    ax.set_title(f'{currency} Exchange Rate Over Time')
    ax.set_xlabel('Date')
    ax.set_ylabel('Exchange Rate')
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))

currencies = ['Chinese Yuan', 'EURO', 'Korean Won', 'UK Pound', 'US Dollar']
for currency in currencies:
    plt.plot(df_filled.index, df_filled[currency], label=currency)

plt.title('Exchange Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.show()

df_returns = df_filled.pct_change().dropna()

plt.figure(figsize=(14, 8))
for currency in currencies:
    plt.plot(df_returns.index, df_returns[currency], label=currency)

plt.title('Daily Percentage Change in Exchange Rates')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.legend()
plt.grid(True)
plt.show()
```

```python
window_size = 30  # 30-day rolling window

plt.figure(figsize=(12, 8))
for currency, color in zip(currencies, colors):
    rolling_volatility = df_filled[currency].pct_change().rolling(window=window_size).std()
    plt.plot(df_filled.index, rolling_volatility, color=color, label=f'{currency} Volatility')

plt.title(f'Rolling {window_size}-Day Volatility of Exchange Rates')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate daily returns
returns = df_filled.pct_change().dropna()

# Calculate correlation matrix
correlation_matrix = returns.corr()

# Visualize correlation matrix using a heatmap
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='summer', linewidths=0.5)
plt.title('Correlation Matrix of Currency Returns')
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Perform PCA on the returns data
pca = PCA(n_components=2)  # We choose 2 components for visualization
pca_result = pca.fit_transform(returns)

# Visualize PCA result
plt.figure(figsize=(10, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Currency Returns')
plt.grid(True)
plt.show()

# Explain variance ratios
print('Explained variance ratios:', pca.explained_variance_ratio_)
```

```python
from statsmodels.tsa.api import VAR

# Fit VAR model
model = VAR(returns)
results = model.fit(maxlags=15)

# Impulse Response Function
irf = results.irf(10)
irf.plot(orth=True)
plt.show()
```

```python
from statsmodels.tsa.regime_switching.markov_regression import MarkovRegression

df_filled = df_filled.dropna()
# Fit Markov Switching Model
model = MarkovRegression(df_filled['Korean Won'], k_regimes=2)
result = model.fit()

# Plot the probabilities of being in each regime
plt.figure(figsize=(10, 6))
plt.plot(result.smoothed_marginal_probabilities[0])
plt.title('Regime 1 Probability (Stable Regime)')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(df_filled['Korean Won'], order=(5, 1, 0))  # Adjust the order parameters as needed
result = model.fit()

# Plot the fitted values
plt.figure(figsize=(10, 6))
plt.plot(df_filled.index, df_filled['Korean Won'], label='Original')
plt.plot(df_filled.index, result.fittedvalues, label='Fitted', color='red')
plt.title('ARIMA Model - Fitted vs. Original')
plt.legend()
plt.show()

```

```python
from arch import arch_model

df_filled = df_filled.dropna()

# Verify if there are any remaining NaN or infinite values
print(df_filled['Korean Won'].isnull().sum())  # Check for NaN values
print(np.isinf(df_filled['Korean Won']).sum())  # Check for infinite values

# Fit GARCH model if no NaN or inf values are present
if df_filled['Korean Won'].isnull().sum() == 0 and np.isinf(df_filled['Korean Won']).sum() == 0:
    model = arch_model(df_filled['Korean Won'], vol='Garch', p=1, q=1)
    result = model.fit()

    # Plot the volatility
    plt.figure(figsize=(10, 6))
    plt.plot(result.conditional_volatility, label='Conditional Volatility')
    plt.title('GARCH Model - Conditional Volatility Over Time')
    plt.show()
else:
    print("There are still NaN or infinite values in the data. Further cleaning is required.")
```

```python
from prophet import Prophet

df_filled_reset = df_filled.reset_index()

# Prepare the data for Prophet
prophet_df = df_filled_reset[['Date', 'Korean Won']].rename(columns={'Date': 'ds', 'Korean Won': 'y'})

# Initialize and fit the Prophet model
model = Prophet()
model.fit(prophet_df)

# Create a future DataFrame for forecasting
future = model.make_future_dataframe(periods=365)

# Make the forecast
forecast = model.predict(future)

# Plot the forecast
model.plot(forecast)
plt.title('Prophet Model - Forecast of Korean Won Exchange Rate')
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit Exponential Smoothing model
model = ExponentialSmoothing(df_filled['Korean Won'], trend='add', seasonal=None, seasonal_periods=12)
result = model.fit()

# Plot the original vs fitted values
plt.figure(figsize=(10, 6))
plt.plot(df_filled.index, df_filled['Korean Won'], label='Original')
plt.plot(df_filled.index, result.fittedvalues, label='Fitted', color='red')
plt.title('Exponential Smoothing - Fitted vs. Original')
plt.legend()
plt.show()
```
