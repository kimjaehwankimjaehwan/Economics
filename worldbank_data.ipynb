```python
import pandas as pd

df = pd.read_excel('/content/P_Popular Indicators.xlsx')
df.head()
```

```python
df1 = df[df['Series Name']=='Inflation, consumer prices (annual %)']
```

```python
df1.head()
```

```python
df1 = pd.DataFrame(df1)
df1.drop(['Series Name','Series Code','Country Code'],axis=1,inplace=True)
```

```python
df1.reset_index(inplace=True, drop=True)
df1.head()
```

```python
df1.columns = ['Country', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009','2010', '2011', '2012', '2013', '2014', '2015']
```

```python
df1.head()
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(12, 8))

for col in df1.columns[1:]:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')

ax = df1.plot(x='Country', kind='bar', stacked=True, ax=plt.gca())

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.xlabel('Country')
plt.ylabel('Inflation')
plt.title('Inflation by Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df2 = df[df['Series Name']=='GDP growth (annual %)']
```

```python
df2.head()
```

```python
df2 = pd.DataFrame(df2)
df2.drop(['Series Name','Series Code','Country Code'],axis=1,inplace=True)
df2.reset_index(inplace=True, drop=True)
df2.columns = ['Country', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009','2010', '2011', '2012', '2013', '2014', '2015']
```

```python
df2
```

```python
df2.drop(df2.index[8], inplace=True)
```

```python
df2
```

```python
df2.reset_index(inplace=True, drop=True)
```

```python
import numpy as np

plt.figure(figsize=(12, 8))

for country in df2['Country'].unique():
    country_data = df2[df2['Country'] == country].T # Transpose the data
    # Drop the country name from the index to use it as a label
    country_data = country_data.iloc[1:]
    # Convert index to list and values to numeric type for plotting
    plt.plot(country_data.index.tolist(), country_data.values.astype(float), marker='o', label=country)

plt.title('GDP Growth (annual %) Over Time')
plt.xlabel('Year')
plt.ylabel('GDP Growth (annual %)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.grid(True)
plt.show()
```

```python
df3 = df[df['Series Name']=='Income share held by lowest 20%']
df3 = pd.DataFrame(df3)
df3.drop(['Series Name','Series Code','Country Code'],axis=1,inplace=True)
df3.reset_index(inplace=True, drop=True)
df3.columns = ['Country', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009','2010', '2011', '2012', '2013', '2014', '2015']
```

```python
df3
```

```python
df3.replace('..', np.nan, inplace=True)
```

```python
df3
```

```python
# 결측치를 각 국가별 평균값으로 대체
df3_filled = df3.set_index('Country').apply(lambda x: x.fillna(x.mean()), axis=1)

print(df3_filled)
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(df3_filled, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Income Share (%)'})
plt.title('Income Share Held by Lowest 20% (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for country in df3_filled.index:
    plt.plot(df3_filled.columns, df3_filled.loc[country], marker='o', label=country)

plt.title('Income Share Held by Lowest 20% Over Time')
plt.xlabel('Year')
plt.ylabel('Income Share (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df3_filled.transpose().plot(kind='area', stacked=True, alpha=0.5)

plt.title('Income Share Held by Lowest 20% Over Time (Stacked Area)')
plt.xlabel('Year')
plt.ylabel('Income Share (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df3_filled.T.plot(kind='box')

plt.title('Distribution of Income Share Held by Lowest 20%')
plt.xlabel('Year')
plt.ylabel('Income Share (%)')
plt.grid(True)
plt.show()
```

```python
from math import pi

labels = df3_filled.columns
num_vars = len(labels)

plt.figure(figsize=(12, 8))
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

ax = plt.subplot(111, polar=True)

for country in df3_filled.index:
    values = df3_filled.loc[country].tolist()
    values += values[:1]
    ax.plot(angles, values, marker='o', label=country)
    ax.fill(angles, values, alpha=0.1)

plt.title('Income Share Held by Lowest 20% (Radar Chart)')
plt.xticks(angles[:-1], labels)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
